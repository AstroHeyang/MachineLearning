### 3.2 Gradient Descent
In this part, you will fit the linear regression parameters $\theta$ to our dataset using gradient descent.

#### Update Equations
The objective of linear regression is to minimize the cost function

$$J(\theta)=\frac{1}{2m} \displaystyle \sum_{i=1}^{m}(h_\theta(x^{(i)})-y^{(i)})^2 \tag{1} $$   

where the hypothesis $h_\theta(x)$ is given by the linear model

$$h_\theta(x)=\theta^T=\theta_0+\theta_1x_1 \tag{2}$$   

Recall that the parameters of your model are the θj values. These are the values you will adjust to minimize cost $J(\theta)$. One way to do this is to use the batch gradient descent algorithm. In batch gradient descent, each iteration performs the update

$$\theta_j:=\theta_j-\alpha\frac{1}{m} \displaystyle\sum_{i=1}^{m}(h_\theta(x^{(i)})-y^{(i)})x_j^{(i)} \tag{3}$$ 
                                                                                                                       

```python
# 获取X数据
# data_x = data1['populations']
ones = pd.DataFrame({'ones': np.ones(len(data1))})
data = pd.concat([ones, data1], axis=1)
data_x = np.array(data.iloc[:, :-1])
```

```python
# 获取Y数据
data_y = np.array(data1['profit'])
```

```python
# 初始值
theta = np.zeros(data_x.shape[1])
alpha = 0.01
iterations = 1500 # 最大迭代次数
precision = 1e-3
theta
```

```python
# 计算损失函数(cost function)初始值, 定义见公式(1)
def calc_cost(theta, data_x, data_y):
    m = data_x.shape[0]
    inner = data_x @ theta - data_y # @ = dot product
    cost = inner.T @ inner / (2 * m) # 平方 = 矩阵点乘其转置矩阵
    return cost
cost = calc_cost(theta, data_x, data_y)
cost
```

```python
# 计算梯度(gradient),定义见公式(3)后半部分
def gradient(theta, data_x, data_y):
    m = data_x.shape[0]
    inner = data_x.T @ (data_x @ theta - data_y) # (m,n).T @ (m, 1) -> (n, 1)
    return inner/m
gradient_init = gradient(theta, data_x, data_y)
gradient_init[0] 
```

```python
# 批量梯度下降 batch gradient decent 
def batch_gradient_decent(theta, alpha, data_x, data_y, iterations):
    # 返回线性回归值，以及代价函数的值列表，用于记录代价函数值的变化
    cost_data = [calc_cost(theta, data_x, data_y)] 
    _theta = theta.copy()
    theta_data = [_theta]
    for _ in range(iterations):
        gradient_current = gradient(_theta, data_x, data_y)
        _theta = _theta - alpha * gradient_current
        theta_data.append(_theta)
        cost_data.append(calc_cost(_theta, data_x, data_y))
    # print(gradient_current)
    return theta_data, cost_data
```

```python
theta_data, cost_data = batch_gradient_decent(theta, alpha, data_x, data_y, iterations)
theta_final = theta_data[-1]
cost_data[-1]
df_theta = pd.DataFrame(theta_data, columns = ["theta0", "theta1"])
df_theta.head()
np.reshape(theta_data, (2,))
```

### 3.3 Visualizing $J(\theta)$
To understand the cost function $J(\theta)$ better, you will now plot the cost over a 2-dimensional grid of $\theta_0$ and $\theta_1$ values

```python
# from matplotlib.mlab import bivariate_normal
sns.lineplot(x="iterations", y="cost",
             data=pd.DataFrame({"iterations":list(range(len(cost_data))), "cost":cost_data}))
# ax = sns.tsplot(cost_data, time=np.arange(iterations+1))
plt.show()
#可以看到第二次迭代的cost变化幅度最大，之后的变化趋于平稳
```

### 4.1 数据归一化
By looking at the values, note that house sizes are about 1000 times the number of bedrooms. When features differ by orders of magnitude, first performing feature scaling can make gradient descent converge much more quickly. (数据归一化是为了加快收敛速度)


- item Subtract the mean value of each feature from the dataset. (第一步减去平均值)
- After subtracting the mean, additionally scale (divide) the feature values by their respective “standard deviations.” （第二步再除以标准差）

```python
def normalize_feature(df):
    return df.apply(lambda column: (column - column.mean()) / column.std() )
data2_norm = normalize_feature(data2)
print(data2_norm.info())
data2_norm.head()
```

### 4.4 Normal Equations (正规方程)
In the lecture videos, you learned that the closed-form solution to linear regression is
$$\theta=(X^TX)^{-1}X^T\vec{y} \tag{4}$$
Using this formula does not require any feature scaling, and you will get an exact solution in one calculation: there is no “loop until convergence” like in gradient descent.
正规方程不需要进行多次迭代，一次就可以求解得到，但是需要求逆矩阵，时间复杂度为O(n^3)，适用于数据量不大的线性模型。

```python
def normal_equation(X, Y):
    theta = np.linalg.inv(X.T@X)@X.T@Y
    return theta
```

```python
theta_final_norm_eq = normal_equation(data_x, data_y) #与批量梯度下降的theta的值略有不同
print(theta_final)
print(theta_final_norm_eq)
```
